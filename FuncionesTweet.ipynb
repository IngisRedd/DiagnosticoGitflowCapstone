# Importar librer√≠as y leer datos

```python
import json
import re
```

```python
def read_data(read_path):
  tweets = []
  with open(read_path, "r") as file:
    for line in file:
      tweets.append(json.loads(line))
  return tweets
```

# Funci√≥n top 10 retweets

```python
def find_top_10_retweets(tweet_list):
  if len(tweet_list) < 10:
    print("not enough tweets")
    return
  
  top_10_stack = [None] * 10
  opponent = top_10_stack[-1]

  for tweet in tweet_list:
    opponent_idx = 9
    if not opponent or tweet["retweetCount"] > opponent["retweetCount"]:
      # print()
      top_10_stack[opponent_idx] = tweet
      opponent_idx -= 1
      opponent = top_10_stack[opponent_idx]
      while opponent_idx > -1 and (not opponent or tweet["retweetCount"] > opponent["retweetCount"]):
        top_10_stack[opponent_idx] = tweet
        top_10_stack[opponent_idx + 1] = opponent
        opponent_idx -= 1
        opponent = top_10_stack[opponent_idx]
        # print([x["retweetCount"] for x in top_10_stack if x])
  return top_10_stack
```

# Funci√≥n main

```python
def main():
  # tweets = read_data("/content/farmers-protest-tweets-2021-03-5.json")
  top_10_re_t = find_top_10_retweets(tweets)
  for tw in top_10_re_t:
    print(tw)
    print(tw["retweetCount"])
    print()
```
