# Importar librerías y leer datos

```python
import json
import re
```

```python
def read_data(read_path):
  tweets = []
  with open(read_path, "r") as file:
    for line in file:
      tweets.append(json.loads(line))
  return tweets
```

# Función top 10 tweets + retweeteados

```python
def find_top_10_retweets(tweet_list):
  return sorted(tweet_list, key=lambda x: x['retweetCount'], reverse=True)[:10]
```

# Función top 10 usuarios que + tweets emiten

```python
def find_top_10_users(tweet_list):
  users = dict()
  for tweet in tweet_list:
    if tweet["user"]["username"] not in users:
      users[ tweet["user"]["username"] ] = {"user":tweet["user"], "ntweets":1}
    else:
      users[ tweet["user"]["username"] ]["ntweets"] += 1
  sorted_users = sorted(users.values(), key=lambda x: x['ntweets'], reverse=True)[:10]
  return [x["user"] for x in sorted_users]
```

# Función top 10 días con + tweets

```python
def find_top_10_days(tweet_list):
  days = dict()
  for tweet in tweet_list:
    if tweet["date"][:10] not in days:
      days[ tweet["date"][:10] ] = 1
    else:
      days[ tweet["date"][:10] ] += 1
  return sorted(days, key=days.get, reverse=True)[:10]
```

# Función top 10 hastags + usados

```python
# Utilizando código recuperado de:
# https://www.geeksforgeeks.org/python-extract-hashtags-from-text/

def extract_hashtags(text):
	
	# the regular expression
	regex = "#(\w+)"
	
	# extracting the hashtags
	hashtag_list = re.findall(regex, text)
	
	return hashtag_list

def find_top_10_hastags(tweet_list):
  hashtags = dict()
  for tweet in tweet_list:
    hashtag_list = extract_hashtags(tweet["content"])
    for hashtag in hashtag_list:
      if hashtag not in hashtags:
        hashtags[ hashtag ] = 1
      else:
        hashtags[ hashtag ] += 1
  return sorted(hashtags, key=hashtags.get, reverse=True)[:10]
```
